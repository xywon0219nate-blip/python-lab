# 문자열

- 문자열 : 파이썬에서는 따옴표("",'')로 둘러싼 문자 집합을 의미

## 1. 문자열 생성
- 시작과 끝이 같은 따옴으로 사용한다, ("@', '@" 와 같은 형식 불가)

```python
str1 = 'hello_'
str2 = "python"
print(str1+str2)
print(type(str1), type(str2))
```

```python
str3 = '안녕하세요 저는 "정예원" 입니다'
print(str3)
```

## 2. 문자열 다루기

- 더하기 연산자(+)와 곱하기 연산자(*) 이용
- (+) 연산자 : 문자열 결합(concatenation)을 통해 하나로 생성함
- (*) 연산자 : 곱한 수만큼 문자열 반복됨

```python
a = '안녕'
b = '하세요'
c = a+b
c
```
