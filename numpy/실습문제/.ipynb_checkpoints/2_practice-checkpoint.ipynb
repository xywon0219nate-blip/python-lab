## 실습 4 정답

````python
import numpy as np

scores = np.array([55, 72, 88, 43, 95, 61, 79, 82, 90, 67])

print("1번:", scores[scores >= 70])
print("2번:", scores[(scores < 60) | (scores >= 90)])

filtered = scores[(scores >= 70) & (scores < 90)]
print("3번:", np.mean(filtered))

print("4번:", np.sum(scores >= 80))
````

## 실습 5 정답

````python
import numpy as np

products = np.array(['사과', '바나나', '포도', '딸기', '오렌지'])
prices   = np.array([1500, 800, 3000, 2500, 1200])

sort_idx = np.argsort(prices)
print("1번 (가격 오름차순):", products[sort_idx])

desc_idx = np.argsort(prices)[::-1]
print("2번 상품:", products[desc_idx])
print("2번 가격:", prices[desc_idx])

most_expensive_idx = np.argsort(prices)[-1]
print("3번 상품:", products[most_expensive_idx])
print("3번 가격:", prices[most_expensive_idx])
````

## 실습 6 정답

````python
import numpy as np

names  = np.array(['김민준', '이서연', '박지호', '최수아', '정우진',
                   '강예린', '윤도현', '임하은', '오준서', '한채원'])
scores = np.array([85, 92, 78, 95, 63, 88, 71, 99, 56, 82])

avg  = np.mean(scores)
high = np.max(scores)
low  = np.min(scores)
print(f"평균: {avg}, 최고점: {high}, 최저점: {low}")

above_avg = names[scores >= avg]
print("평균 이상:", above_avg)

top3_idx = np.argsort(scores)[-3:][::-1]
print("상위 3명 이름:", names[top3_idx])
print("상위 3명 점수:", scores[top3_idx])

print("60점대:", np.sum((scores >= 60) & (scores < 70)))
print("70점대:", np.sum((scores >= 70) & (scores < 80)))
print("80점대:", np.sum((scores >= 80) & (scores < 90)))
print("90점 이상:", np.sum(scores >= 90))
````
