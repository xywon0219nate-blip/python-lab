# 배열 연산
- 배열의 shape이 같으면 데이터별로 연산이 진행됨

```python
import numpy as np
```

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
print(arr1.shape, arr2.shape)
print(arr1, arr2)
```

```python
print(arr1 + arr2)
print(arr1 - arr2)
print(arr1 * arr2)
print(arr1 / arr2)
```

```python
# shape이 다른 경우 ValueError 발생 💥
arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([4, 5, 6])
print(arr1.shape, arr2.shape)
print(arr1 + arr2)
```

- 스칼라(단일 숫자)와 연산

```python
print(arr1 + 10)
print(arr1 - 1)
print(arr1 * 3)
print(arr1 / 2)
```

- 2차원 배열 연산

```python
arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[5, 6], [7, 8]])
print(arr1.shape, arr2.shape)
```

```python
print(arr1 + arr2)
print(arr1 - arr2)
print(arr1 * arr2)
print(arr1 / arr2)
```

# 정렬 - sort()와 argsort()

### (1) sort()
- numpy.sort() vs numpy.ndarray.sort() : 원본 배열의 유지 여부가 핵심
- 원본 유지 ⭕ : numpy.sort()
- 원본 유지 ❌ : numpy.ndarray.sort()
- numpy.sort() 사용을 권장함!!

```python
org_arr = np.array([2, 1, 3, 6, 9])
org_arr
```

```python
# 오름 차순 정렬 후 새 배열 반환
sort_arr = np.sort(org_arr)
sort_arr
```

```python
print(org_arr, sort_arr)
```

```python
# 원본 배열을 직접 오름차순 정렬
org_arr.sort()
print(org_arr)
```

- 내림차순 정렬 : 배열[::-1]

```python
org_arr = np.array([8, 4, 9, 2, 5])
sort_arr = np.sort(org_arr)[::-1] # sort 실행 반환 결과 > 슬라이싱
sort_arr
```

```python
org_arr = np.array([8, 4, 9, 2, 5])
org_arr.sort() # sort 실행 반환 결과 None > 슬라이싱 불가
org_arr[::-1]
```

- 2차원 배열 정렬

```python
arr = np.array([[8, 12], 
               [7, 1]])
arr
```

```python
sort_arr = np.sort(arr, axis=0) # axis = 0 행 기준으로 정렬
sort_arr
```

```python
sort_arr2 = np.sort(sort_arr, axis=1) # axis = 1 열 기준으로 정렬
sort_arr2
```

### (2) argsort()
- 정렬된 인덱스 반환

```python
names = np.array(['홍길동','김유신','강감찬','이순신', '스미스'])
scores = np.array([72, 81, 95, 66, 88])
```

```python
# 성적순으로 names 배열을 오름차순 정렬
sort_idx = np.argsort(scores)  # 인덱스 반환 : [3, 0, 1, 4, 2]
sort_idx
```

```python
# 정렬한 인덱스를 팬시 인덱싱 실행
names[sort_idx]
```

```python
# 성적순으로 names 배열을 내림차순 정렬
sort_idx2 = np.argsort(scores)[::-1]
names[sort_idx2]
```

# 집계 함수

```python
arr = np.arange(1, 10)
arr
```

```python
print(np.sum(arr)) # 합계
print(np.mean(arr)) # 평균
print(np.max(arr)) # 최대값
print(np.min(arr)) # 최소값
print(np.std(arr)) # 표준편차
print(np.var(arr)) # 분산
print(np.median(arr)) # 중앙값
```

```python
arr = np.array([2, 3, 1, 20, 7, 8])
print(np.sum(arr))
print(np.mean(arr))    # 평균
print(np.median(arr))  # 중앙값 [1, 2, 3, 7, 8, 20] : 오름차순 정렬 후 짝수인 경우 중앙의 2값을 더한후 2로 나눈값
```

- 2차원 배열에서 집계함수 사용

```python
arr = np.array([[1, 2, 3],
                [4, 5, 6]])
print(f"전체 합계 : {np.sum(arr)}")
print(f"행기준 합계 : {np.sum(arr, axis=0)}")
print(f"열기준 합계 : {np.sum(arr, axis=1)}")

print(f"행기준 평균 : {np.mean(arr, axis=0)}")
print(f"열기준 평균 : {np.mean(arr, axis=1)}")
```

# 선형 대수 연산

### (1) 형렬 내적(dot product)
- 행렬의 곱셈은 dot 함수 사용
  예) A(2,3) * B(3,2) = C(2,2)
- 규칙 : A의 열 수 === B의 행 수가 동일할때만 내적 연산이 가능함  

```python
A = np.array([[1, 2, 3],
              [4, 5, 6]])
B = np.array([[7, 8], 
              [9, 10], 
              [11, 12]])
C = np.dot(A, B)
print(f"A : {A}")
print(f"B : {B}")
print(f"C : {C}")
```

### (2) 전치 행렬
- np.transpose() 함수를 이용하여 행과 열을 바꿈

```python
# A 배열의 전치 행렬을 구하기
A_T = np.transpose(A)
print(A.shape)
print(A_T.shape)
print(f"A : {A}")
print(f"A_T : {A_T}")
```

# 기타 함수

### (1) np.where()
- 조건에 따라 값을 선택함

```python
arr = np.arange(1, 10)
arr
```

```python
# 5보다 큰 값만 출력
arr2 = arr[arr > 5]
arr2
```

```python
# 5보다 큰 값만 출력
idx = np.where(arr>5)
print(idx)
arr2 = arr[idx] # [1, 2, 3, 4, 5, 6, 7, 8, 9]
print(arr2)
```

### (2) np.unique()
- 중복없는 고유한 값을 반환

```python
arr = np.array([1, 2, 3, 3, 2, 1, 4, 5])
print(np.unique(arr))
```

```python
# value, count 출력
values, counts = np.unique(arr, return_counts=True)
print(values)
print(counts)
```

### (3) np.concatenate()
- 배열을 이어붙이기

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
c = np.concatenate([a, b])
c
```
