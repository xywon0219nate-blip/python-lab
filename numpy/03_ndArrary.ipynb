# 1. ndArray

### (1) ndArray 생성

- 형식 : numpy.array() 함수에 객체를 인자로 추가하여 생성한다.
- 객체로 파이썬의 리스트를 생성하여 입력하여 자동으로 변환된다.

```python
import numpy as np

# 1차원 배열 생성
# 리스트를 ndarray로 변환하여 생성
list1 = [1, 2, 3]
array1 = np.array(list1)
print(type(list1), type(array1))

array2 = np.array([1, 2, 3])
print(type(array2))
```

```python
array2 = np.array([1, 2, 3])
print(type(array2))
```

```python
# 튜플을 ndarray로 변환하여 생성
tuple1 = (1, 2, 3)
tarray1 = np.array(tuple1)
print(type(tarray1))
print(tarray1)
```

```python
# 세트를 ndarray로 변환하여 생성
set1 = {1, 2, 3}
sarray1 = np.array(list(set1))
print(type(sarray1))
print(sarray1)
```

```python
# 2차원 배열 생성
list2 = [[1,2],[3,4]]
array2 = np.array(list2) # 중첩 리스트 사용
print(type(array2))
print(array2)
```

```python
# 3차원 배열 생성
list3 = [[[1,2]],[[3,4]]]
array3 = np.array(list3) # 중첩 리스트 사용
print(type(array3))
print(array3)
```

### (2) shape와 ndim 
- shape : 배열의 모양(형태)
- ndim : 배열의 차원

```python
array1 = np.array([1, 2, 3])
array2 = np.array([[1], [2], [3]])
array3 = np.array([[[1]], [[2]], [[3]]])
```

```python
print(f"1차원 : {array1.shape}")
print(f"2차원 : {array2.shape}")
print(f"3차원 : {array3.shape}")
```

```python
print(f"1차원 : {array1.ndim}")
print(f"2차원 : {array2.ndim}")
print(f"3차원 : {array3.ndim}")
```

### (3) dtype 
- 저장된 데이터 타입 확인
- 모든 데이터 타입이 동일 해야함

```python
# 정수 배열
arr_int = np.array([1, 2, 3])
print(type(arr_int))
print(arr_int.dtype) # int64
```

```python
# 실수 배열
arr_float = np.array([1.23, 2.34, 3.0])
print(type(arr_float))
print(arr_float.dtype) # float64
```

```python
# 문자열 배열
arr_str = np.array(['a','b','c'])
print(type(arr_str))
print(arr_str.dtype) # <U1
```

### (4) 자동 형변환

- 여러 타입이 저장되는 경우 가장 큰 타입으로 자동 변환

```python
# 정수 + 문자열 => 문자열 타입 저장
arr = np.array([1, 2, 'abc'])
arr
```

```python
# 정수 + 실수 => 실수 타입으로 저장
arr = np.array([1, 2, 3.123])
arr
```

```python
# 실수 + 문자열 => 문자열 타입으로 저장 : 정수 < 실수 < 문자열
arr = np.array([1.0, 2.12, 'test'])
arr
```

### (5) 데이터 타입 변환 - astype()
- 데이터 타입을 명시적으로 변환
- 자주 사용하는 dtype 종류
  |dtype|설명|메모리|
  |------|----|-----|
  |int32|32비트 정수|4바이트|
  |int64|64비트 정수(기본값)|8바이트|
  |float32|32비트 실수|4바이트|
  |float64|64비트 실수(기본값)|8바이트|
  |bool|불린 (True/False)|1바이트|
  |<U숫자|유니코드 문자열|가변|

```python
arr_int = np.array([1, 2, 3])
arr_int.dtype
```

```python
# arr_int 배열을 float64 타입의 배열로 변환
arr_float = arr_int.astype('float64')
arr_float
```

```python
# arr_float 배열을 int32 타입의 배열로 변환
arr_int32 = arr_float.astype('int32')
arr_int32
```

```python
# arr_int32 배열을 문자열 배열로 변환
arr_str = arr_int32.astype('<U1')
arr_str
```

### (7) np.zeros(), np.ones()
- np.zeros() : 모든 데이터가 0인 배열
- np.ones() : 모든 데이터가 1인 배열

```python
# 3행의 1차원 배열 생성 후, 모든 데이터를 0으로 초기화
arr_zeros = np.zeros((3,), dtype='int64')
arr_zeros
```

```python
# 모든 데이터를 0으로 갖는 3행 4열의 2차원 배열 생성
arr_zeros = np.zeros((3, 4), dtype='int64')
arr_zeros
```

```python
# 모든 원소를 정수 1로 가지는 5행의 1차원 배열 생성
arr_ones = np.ones((5,), dtype='int64')
arr_ones
```

```python
# 모든 원소를 실수 1로 가지는 2행 5열의 2차원 배열 생성
arr_ones = np.ones((2, 5), dtype='float64')
arr_ones
```

### (9) np.random
- 0과 1사이의 난수를 발생
- rand(shape) 함수를 이용하여 실수형 배열 생성
- randint(high, size = shape) 함수를 이용하여 정수형 배열 생성, high 값보다 작은 정수만 저장

```python
# 난수를 발생하여 3행의 1차원 배열 생성
arr_random = np.random.rand(3,)
arr_random
```

```python
# 난수를 발생하여 5행 4열의 2차원 배열 생성
arr_random = np.random.rand(5,4)
arr_random
```

```python
# 난수를 발생하여 5보다 작은 정수들만, 3행의 1차원 배열 생성
arr_random = np.random.randint(5, size=(3,))
arr_random
```

```python
# 난수를 발생하여 10보다 작은 정수들만, 3행 5열의 2차원 배열 생성
arr_random = np.random.randint(10, size=(3, 5))
arr_random
```

### (10) axis - 배열의 축
- 2차원 배열의 axis=0은 row(행), axis=1 column(컬럼) 방향을 의미

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])
arr.shape
```

```python
# 행(가로) 기준으로 합계 계산
arr_row_sum = np.sum(arr, axis=0)
print(arr_row_sum)
# 1 2 3
# 4 5 6
```

```python
# 컬럼(세로) 기준으로 합계 계산
arr_col_sum = np.sum(arr, axis=1)
print(arr_col_sum)
```

### (11) reshape()
- 배열의 형태를 바꾸는 함수
- 배열의 원소 개수를 유지하면서 변환해야 함: 데이터가 남거나 부족하면 Value Error 발생!!

```python
arr = np.arange(10)
arr
```

```python
# 1차원 arr 객체를 2행 5열의 2차원 배열로 변환
arr2 = np.reshape(arr, (2,5))
print(arr2.shape)
print(arr2)
```

```python
# 2행 5열의 2차원 배열을 1차원 배열로 변환
arr3 = np.reshape(arr2, (10,))
print(arr3.shape)
print(arr3)
```

```python
arr4 = np.arange(12).reshape((3,4))
print(arr4.shape)
print(arr4)
```

```python
arr5 = np.reshape(arr4, (2,6))
print(arr5.shape)
print(arr5)
```

- -1을 사용하여 자동 계산

```python
arr = np.arange(10)
arr
```

```python
# 5열의 2차원 배열로 생성 - 컬럼은 고정, 행은 자동계산(-1)
arr2 = np.reshape(arr, (-1, 5))
print(arr2.shape)
print(arr2)
```

```python
# 2행의 2차원 배열로 생성 - 컬럼은 자동계산(-1), 행은 2행 고정
arr3 = np.reshape(arr, (2, -1))
print(arr3.shape)
print(arr3)
```

```python
# 원소 갯수 256개  2컬럼을 가진 2차원 배열 생성
arr4 = np.arange(256).reshape((-1, 2))
arr4.shape
```

```python
# 8개 원소를 가진 1차원 배열 => 3차원 배열로 변환
arr_3d = np.arange(8).reshape((2, 2, 2))
print(arr_3d.shape)
print(arr_3d)
```

```python
# 3차원 배열 => 1차원 변환
arr_1d = np.reshape(arr_3d, (8,))
print(arr_1d.shape)
print(arr_1d)
```
