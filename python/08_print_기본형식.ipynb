# print()

## 1. 기본 문법
- print(value, ..., sep='구분기호', end='\n', flush=False) 

```python
a = 5
a
b = 3
b
a + b
```

```python
print(a)
print(b)
print(a+b)
```

```python
print(1, 2, 3, sep='-')  # 1-2-3
```

```python
print(1, 2, 3, sep='-', flush=True)  # 1-2-3
```

```python
print(a, end='\t')
print(b, end='\t')
print(a+b)
```

```python
name = '홍길동'
age = 30
```

```python
# 1. 콤마(,) 로 구분
print('이름:', name, '나이:', age) 
```

```python
# 2. 접합연산자(+)를 사용하여 문자열 연결, 문자열만 가능
print('이름:'+ name, '나이:'+ str(age)) # int()-숫자로 변환, str()-문자열로 변환
```

```python
# 3. % 포맷 (old version)
print("이름: %s, 나이: %d" % (name, age))
```

```python
# 4. format{} 포맷 
print("이름: {}, 나이: {}".format(name, age))
```

```python
# 5. f-string (권장)
print(f"이름: {name}, 나이: {age}")
```

```python
# 여러 값을 출력 콤마(,) 구분
print(1, 2, 3)
print(1, 2, 3, sep=',')
```

```python
# 줄바꿈 없이 옆으로 출력
lst = [1, 2, 3, 4, 5]
for i in lst:   
    print(i, end='\t')
    
```

```python
# 빈 줄 출력
print()
```

```python
# '*' 문자열을 50번 출력
print('*' * 50)
```
