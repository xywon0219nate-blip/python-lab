# 딕셔너리

- 키(key), 값(value)의 쌍으로 구성된 데이터 타입
- 키(key) : 임의로 지정한 숫자나 문자열
- 값(value): 어떤 데이터 타입도 가능
- 중괄호 {키:값, 키:값, ... }

## 1. 딕셔너리 생성

```python
dict1 = {
    'name': '홍길동',
    'age': 30,
    'address': '서울시 강남구'
}
```

```python
dict1
```

```python
fruits = {
    "name": "apple",
    "color": "red",
    "emoji": "🍎"
}
```

```python
fruits
```

## 2. 아이템 선택/추가
- 딕셔너리 key를 이용한 데이터 변경
- 추가/수정 : update()

```python
# 딕셔너리 객체의 모든 키 리턴
dict1_keys = dict1.keys()
```

```python
type(dict1_keys)
dict1_keys
```

```python
fruits.keys()
```

```python
fruits['emoji']
```

```python
# 딕셔너리 key를 이용한 데이터 변경
fruits['emoji'] = '🍏'
fruits
```

```python
# 딕셔너리 update함수를 이용한 데이터 변경 - {key:value}
fruits.update({"color":"green"})
fruits
```

```python
# 딕셔너리에 데이터 추가, key가 있으면 update, 없으면 추가
fruits.update({"price": 1000})
fruits
```

```python
# 나이를 20으로 수정, 'job'은 '개발자' 추가
dict1.update({"age":20, "job":"개발자"})
```

```python
dict1
```

## 4. 딕셔너리 메소드

- keys() : 딕셔너리 객체의 모든 키를 리스트 형태로 반환

```python
fruits.keys()
```

- values() : 딕셔너리 객체의 모든 값을 리스트 형태로 반환

```python
fruits.values()
```

- items() : 딕셔너리 객체의 (key:value) 값을 튜플 형태로 반환

```python
fruits.items()
```

- update(item or dict) : 딕셔너리 객체에 아이템 또는 딕셔너리 추가

```python
fruits.pop('emoji')
```

```python
fruits.pop('color')
```

```python
fruits
```

```python
fruits2 = {'name':'orange', 'color':'coral', 'price':2000}
fruits2
```

```python
fruits.update(fruits2)
fruits
```

- clear() : 모든 아이템 삭제

```python
fruits.clear()
```

```python
fruits
```

```python
print(dir())
```

- copy() : 딕셔너리 객체 복사 : JS의 deep copy 형식으로 독립적인 객체로 생성

```python
dict1 = {'korea': '서울', 'japan':'도쿄'}
```

```python
dict2 = dict1.copy() # 메모리 영역이 독립적인 객체로 생성, 원본(dict1)에 영향을 주지 않음
```

```python
print(dict1, dict2)
```

```python
# dict2의 korea 아이템 삭제
dict2.pop('korea')
```

```python
print(dict1, dict2)
```

```python
dict3 = dict1  # JS의 shallow copy 형식으로 주소값을 공유함, 원본에 영향을 줌
```

```python
print(dict1, dict3)
```

```python
dict3.pop('korea')
```

```python
print(dict1, dict3)
```
