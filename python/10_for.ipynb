# 반복문

- 반복적인 작업을 수행하는 구문
- for문, while문

## 1. for문

```python
# 리스트 객체를 반복
for number in [1, 2, 3, 4, 5]:
    print(number, end='\t')
```

```python
# fruits 리스트 객체 생성 및 출력
fruits = ['apple','lemon','mango']
for fruit in fruits:
    print(fruit, end='\t')
```

```python
# range(start, stop, step) 함수를 이용한 리스트 객체 생성
for x in range(0,5,1):  # [0, 1, 2, 3, 4]
    print(x)
```

```python
# range(start, stop, step) 함수를 이용한 리스트 객체 생성
for x in range(5):  # [0, 1, 2, 3, 4]
    print(x+1, end=" ")
```

#### range(start, stop, step)

- start는 범위의 시작 지점, stop은 범위의 끝 지점, step은 증감의 크기
- step이 1인 경우 생략 가능
- step이 1이고 start가 0인 경우 start역시 생략 가능

```python
list1 = list(range(1, 11))
list1
```

```python
print(list(range(10)))
```

```python
print(set(range(10)))
print(tuple(range(10)))
```

```python
for i, fruit in enumerate(fruits):
    print(i, fruit)
```

```python
# 101 ~ 110까지 출력
# enumerate() : list => key/value
for i, num in enumerate(range(101, 111)):
    print(i, num, sep=":")
```

```python
fruits2 = { "apple":"🍎", "lemon":"🍋", "tomato":"🍅"}
for f in fruits2:
    print(f) # 딕셔너리의 키 값 출력
```

```python
fruits2 = { "apple":"🍎", "lemon":"🍋", "tomato":"🍅"}
for key, emoji in fruits2.items():
    print(key, emoji, sep=",") # 딕셔너리의 키 값 출력
```

```python
# 과일명: apple, 이모지: 🍎  출력
fruits2 = { "apple":"🍎", "lemon":"🍋", "tomato":"🍅"}
for key, emoji in fruits2.items():
    print(f"과일명:{key}, 이모지:{emoji}") # 딕셔너리의 키 값 출력
```

## 2. 중첩 for문

```python
n_list1 = [10, 20]
n_list2 = [30, 40]
for i1, x in enumerate(n_list1):
    for i2, y in enumerate(n_list2):
        print(f"({i1}, {i2}),  {x + y}")  # (0,0) 40
```

```python
# 3단 출력
dan = 3
for n in range(1,10):
    print(f"{dan} ✖ {n} = {dan * n}")
```

```python
# 2단 ~ 9단 출력
# 2*1 = 2 ..... 9*1 = 9
# ...
# 2*9 = 18 ..... 9*9 = 81
for n in range(1, 10):
    for dan in range(2, 10):
        print(f"{dan} ✖ {n} = {dan*n}", end="\t")
    print()
```

## 3. while문

- 조건에 따라 반목 여부 결정

```python
# 1 ~ 5 까지 출력
num = 1
while num <=5:
    print(f"number={num}")
    num += 1
```

```python
# 무한루프 발생 주의!!
# while True:
#     print(출력내용)
```

```python
# while문은 True 실행하는 경우 반드시 if로  조건 체크
k = 0
while True:
    k += 1
    if k > 3:
        break
    else:
        print(k)
```

```python
# continue - 해당 차례의 남은 작업을 스킵하고 다음으로 넘어감
# 1 ~ 5까지 출력, 단 2이면 skip
for n in range(1, 6):
    if n == 2:
        continue
    print(n)
```

## 4. 컴프리헨션(comprehension)

```python
# 리스트 컴프리헨션
numbers = [1, 2, 3, 4, 5]
numbers2 = [k **2 for k in numbers]
print(numbers, numbers2)
```
