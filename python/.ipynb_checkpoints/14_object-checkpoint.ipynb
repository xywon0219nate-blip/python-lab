# OOP(Object Oriented Programming : 객체지향 프로그래밍)

- 객체를 만들고 이용할 수 있는 기능을 제공하는 프로그래밍
- Object(객체)
   - 속성과 행위로 구성
   - 속성 : 객체가 가지는 값, 상태 => 변수로 정의
   - 행위 : 객체가 동작하는 기능, 행위 => 함수로 정의
- Object(객체) 생성 방법
  - 클래스를 선언, 속성(변수), 행위(함수)를 정의
  - 클래스는 기본 틀(설계도)와 같이 틀을 만드는 작업
  - 객체는 클래스를 이용하여 생성되며 인스턴스(instance) 라고도 함

#### (1) 클래스 선언
- 기본 구조  

```python
class 클래스명():
   [변수1]  # 클래스 변수
   [변수2]

   # 생성자 함수
   def __init__(self, 인자1, 인자2...):
        <코드 블록>
   ...
   def 함수1(self[, 인자1, 인자2, ··· , 인자n]): 
       <코드 블록>
       ...
   def 함수2(self[, 인자1, 인자2, ··· , 인자n]):
       <코드 블록>
       ...
```       

````python
# Animal 클래스 생성
class Animal():
    className = "Animal" # 클래스변수 = static 변수

    def __init__(self, name, emoji):  # 생성자 함수, 맨처음 호출
        self.name = name
        self.emoji = emoji

    def display(self):        
        print(f"name: {self.name}, emoji:{self.emoji}")
    
````

````python
# Animal 클래스의 인스턴스 생성
dog = Animal("해피", "🐶")
dog.display()
````

````python
cat = Animal("하양이", "😺")
cat.display()
````

````python
# Bicycle 클래스 정의
class Bicycle():
    # 생성자 함수 정의
    def __init__(self, wheelSize, color, name):
        self.wheelSize = wheelSize
        self.color = color
        self.name = name

    def info(self):
        print(f"wheelSize:{self.wheelSize}, color:{self.color}")

    def move(self, speed):
        print(f"{self.name} 속도는 speed: {speed} 입니다.")
        
````

````python
myBicycle = Bicycle(50, "Red", "따릉이")
myBicycle.info()
myBicycle.move(50)
````

#### (2) 클래스 상속

````python
# Animal 클래스의 자식 Dog 클래스 정의
class Dog(Animal):
    className = "Dog"  # 클래스변수 = static변수

    def __init__(self, name, emoji, color):
        super().__init__(name, emoji)
        self.color = color

    def display(self):
        print(f"name:{self.name}, emoji:{self.emoji}, color:{self.color}")

    def setName(self, name):
        self.name = name

    def getName(self):
        return self.name
````

````python
happy = Animal("해피", "🐶")
happy.display()
print(Animal.className)
````

````python
happy = Dog("해피", "🐶", "브라운")
happy.display()
print(Dog.className)
happy.setName("초코")
happy.display()
print(f"우리 강아지 이름은 {happy.getName()} 입니다.")
````

````python
# Animal 클래스의 자식 Cat 클래스 정의
# 속성 : name, emoji, color, age, type
# 속성에 대한 setter/getter 함수 정의
class Cat(Animal):
    def __init__(self, name, emoji, color, age, ctype):
        self.name = name
        self.emoji = emoji
        self.color = color
        self.age = age
        self.ctype = ctype

    def display(self):
        print(f"name:{self.name}, emoji:{self.emoji}, age:{self.age}, ctype:{self.ctype}")

    def setName(self, name): self.name = name;
    def setEmoji(self, emoji): self.emoji = emoji;
    def setColor(self, color): self.color = color;
    def setAge(self, age): self.age = age;
    def setCtype(self, ctype): self.ctype = ctype;

    def getName(self): return self.name;
    def getEmoji(self): return self.emoji;
    def getColor(self): return self.color;
    def getAge(self): return self.age;
    def getCtype(self): return self.ctype;
````

````python
hoya = Cat("호야", "😺", "white", 10, "샴")
hoya.display()
hoya.setCtype("먼치킨")
hoya.setAge(9)
hoya.display()
````
