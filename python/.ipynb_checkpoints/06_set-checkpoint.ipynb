# Set(세트)

- 수학의 집합개념을 구현한 데이터타입
- 데이터에 순서가 없고, 중복해서 사용이 불가
- 교집합, 합집합, 차집합을 구하는 메서드

## 1. 세트 생성

```python
set1 = {1, 2, 3}
print(set1)
set1
```

```python
type(set1)
```

```python
fruits_set = {'Apple', 'Lemon', 'Apple', 'Lemon'}  # 데이터 중복 불가 ❌ 
fruits_set
```

## 2. 세트(집합) 연산

* 교집합
  - set객체.intersection(set객체)
  - &
    
* 합집합
  - set객체.union(set객체)
  - |
    
* 차집합
  - set객체.difference(set객체)


```python
# 세트 정의
set1 = {1, 2, 3}
set2 = {2, 3, 4}
```

```python
# 교집합
print(set1.intersection(set2), set1&set2)  # 교집합 세트를 생성하여 리턴
```

```python
# 합집합
print(set1.union(set2), set1|set2)
```

```python
# 차집합
print(set1.difference(set2), set2.difference(set1))
```

## 3. 아이템 선택

```python
# 세트 => 리스트로 변환 : list()
fruits_list = list(fruits_set)
print(type(fruits_set), type(fruits_list))
```

```python
fruits_list
```

```python
fruits_list[0:2]
```

## 4. 아이템 추가/삭제
- set객체.add(아이템)
- set객체.update(set객체)

```python
# fruits_set에 'Tomato' 추가 : set객체.add(아이템)
fruits_set.add('Tomato')
fruits_set
```

```python
vegetable_set = {'Carrot', 'Onion'}
vegetable_set
```

```python
# fruits_set에 vegetable_set 추가하여 fruits_set이 update됨
# add, update 시 자동으로 정렬됨
fruits_set.update(vegetable_set)
fruits_set
```

## 5. 아이템 삭제
- set객체.remove(아이템) : 아이템 삭제
- set객체.clear() : 아이템 전체 삭제

```python
fruits_set.remove('carrot')
```

```python
fruits_set.clear()
```

```python
fruits_set
```

```python
del fruits_set
```

## 6. 리스트, 튜플, 세트 간 변환 타입
- list(), tuple(), set()

```python
a = [1, 2, 3, 4, 5]
b = (1, 2, 3, 4, 5)
c = {1, 3, 3, 4, 5}
```

```python
print(type(a), type(b), type(c))
```

```python
# a => tuple, set
a_to_tuple = tuple(a)
a_to_tuple
```

```python
# b => list, set
b_to_set = set(b)
b_to_set.remove(4)
```

```python
b_to_set
```

```python
# c => list, tuple
```

```python
c_to_list = list(c)
c_to_list.append(5)
c_to_list
```

```python
c_to_set = set(c_to_list)
c_to_set
```
