# if문

- 지정한 조건에 따라 다르게 분기해 명령을 수행
- 조건의 만족 여부에 따라서 코드 수행 결과가 달라짐
- 단일 조건에 따른 분기 (if )

⭐ 구문 ⭐

- 코드블럭은 들여쓰기로만 구분
- 보통은 한줄에 한 문장, 길어지면 ''로 연결
- 한줄에 여러 문장을 쓸 때 는 ';' 로구분
- 주석은 '#'으로 시작

```python
# 변수 x의 값이 10보다 크거나 같으면 'Pass'를 출력
x = 10
if x >= 10: 
    print('Pass')
```

### 비교 연산자

|비교 연산자|	의미|	활용 예|	설명|
|----------|-------|----------|---------|
|==	       |같다	|a == b   |	a는 b와 같다|
|!=        |같지 않다|	a != b	|a는 b와 같지 않다|
|<	|작다|	a < b|	a는 b보다 작다|
|>	|크다|	a > b|	a는 b보다 크다|
|<=|	작거나 같다|	a <= b|	a는 b보다 작거나 같다|
|>=|	크거나 같다|	a >= b|	a는 b보다 크거나 같다|

### 논리 연산자

논리 연산자	의미	활용 예	설명	
and	논리곱	A and B	A와 B가 모두 참이면 참이고 그 외에는 거짓	
or	논리합	A or B	A와 B 중 하나라도 참이면 참이고 둘 다 거짓이면 거짓	
not	논리부정	not A	A가 참이면 거짓이고 거짓이면 참

```python
# 1. if - 23도 보다 크면 '아이스아메리카노' 출력
today_temp = 25
if today_temp >= 23:
    print('아이스 아메리카노')
```

```python
# 2. if ~ else - 23도 보다 크면 '아이스아메리카노', 낮으면 '따뜻한 아메리카노' 출력
today_temp = 20
if today_temp >= 23:
    print("아이스 아메리카노")
else:
    print("따뜻한 아메리카노")
```

```python
# 3. if ~ elif ~else - 23도 보다 크면 '아이스아메리카노', 0도 이면 '따뜻한 아메리카노', 조건이 만족하지 않으면 '유자차' 출력
today_temp = 0
if today_temp >= 23:
    print("아이스 아메리카노")
elif today_temp == 0:
    print("따뜻한 아메리카노")
else:
    print("유자차")
```

```python
# 4. 중첩 if
weather = "비"
temp_today = 20
if weather == "맑음":
    if temp_today > 0:
        print("아이스 아메리카노")
    elif temp_today == 0:
        print("미지근한 아메리카노")
    else:
        print("따뜻한 아메리카노")
else:
    print("카푸치노")
```
