# 화면 출력

- print() 함수를 이용하여 화면 출력
- 기본 출력 : 출력 형식 지정하지 않고 데이터를 출력하는 방법
- 형식(attribute) 지정 : sep, end .. 속성을 적용하여 다양하게 출력하는 방법

```python
print('Hello~ Python!!')
```

```python
# 이름은 '홍길동'이고 나이는 30살 입니다.
name = "홍길동"
age = 30
print(f"이름은 '{name}'이고 나이는 {age}살 입니다.")
```

# 키보드 입력

- 키보드로 데이터를 입력받기 위해서는 input() 함수 사용
- 형식 : 변수 = input('문자열')
- input() 함수를 통해 입력받은 값은 기본 문자열로 인식

```python
name = input('이름:')
age = input('나이:')
print(f"이름은 '{name}'이고 나이는 {age}살 입니다.")
```

```python
# 구구단을 입력받아 실행
dan = input('단 입력: ')
for n in range(1, 10):
    print(f"{dan} ✖ {n} = {int(dan)*n}")
```

```python
# 구구단 실행 - 시작, 종료 단을 입력받기
start = input('시작: ')
end = input('종료: ')
for n in range(1, 10):  # 행
    for dan in range(int(start), int(end)+1): # 컬럼
        print(f"{dan} ✖ {n} = {dan * n}", end="\t")
    print()        
```
