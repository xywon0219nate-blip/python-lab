# 3. 변수 이름을 붙이는 규칙(2)

### 🎯 에러 발생은 안되지만, 참고해야할 사항

- 대소문자를 구분
- 한 번에 여러개 변수 선언 가능
- 하나의 값을 여러 변수에 저장 가능

```python
# 1. 대소문자를 구분
abc = 100
abc
```

```python
ABC = 200
ABC
```

```python
print(abc, ABC)
```

```python
# 2. 한 번에 여러개 변수 선언 가능
x, y, z = 100, 200, 300
print(x, y, z)
```

```python
name1, name2 = '홍길동', "이순신"
print(name1, name2)
```

```python
# 3. 하나의 값을 여러 변수에 저장 가능
xx = yy = zz = "Apple"
```

```python
print(xx, yy, zz)
```
