# Tuple(튜플)

- 리스트와 유사
- 한 번 입력 한 후에 항목을 ⭐변경할 수 없는⭐ 특징
- 소괄호()를 사용해서 생성 또는 괄호 없이 생성
- 튜플의 항목은 숫자, 문자열, 불린, 리스트, 튜플, 세트, 딕셔너리 등으로 만들 수 있음

## 1. 튜플 생성

```python
tuple1 = (1, 2, 3)
tuple1
```

```python
tuple1[0]
```

```python
tuple1[0:2]
```

```python
tuple2 = (5, 6, 7, 8)
tuple2
```

```python
tuple2[1:]
```

- 인자가 하나만 존재하는 튜플 생성

```python
tuple3 = (9,) # 소괄호 생략 가능 
tuple3
```

## 2. 튜플 아이템 선택

```python
fruits_tuple = ('Apple', 'Tomato');
fruits_tuple
```

```python
type(fruits_tuple)
```

```python
fruits_tuple[1]
```

```python
fruits_tuple[1] = 'Mango' # 아이템 변경 불가 ❌
```

## 3. 튜플 아이템 추가/삭제 - 실행 불가❌

```python
fruits_tuple
```

```python
fruits_tuple.append('Lemon') # 새로운 아이템 추가 안됨 ❌
```

```python
fruits_tuple.remove('Apple') # 아이템 삭제 안됨 ❌
```

## 4. 튜플을 리스트로 형변환

```python
type(fruits_tuple)
```

```python
fruits_list = list(fruits_tuple)
```

```python
type(fruits_list)
```

```python
# fruits_list 마지막에 'orange' 추가
fruits_list.append('orange')
fruits_list
```

```python
fruits_tuple = tuple(fruits_list)
```

```python
fruits_tuple
```

```python
# fruits_tuple 객체의 orange 아이템을 Mango로 변경
f_tuple = ('Apple', 'Orange')
f_list = list(f_tuple)
f_list[f_list.index('Orange')] = 'Mango'
f_tuple = tuple(f_list)
```

```python
print(type(f_tuple), f_tuple)
```

## 5. 튜플 삭제

```python
del fruits_tuple
```
