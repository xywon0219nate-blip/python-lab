# 딕셔너리

- 키(key), 값(value)의 쌍으로 구성된 데이터 타입
- 키(key) :  임의로 지정한 숫자나 문자열
- 값(value): 어떤 데이터 타입도 가능


- 중괄호 {키:값, 키:값,  ... }

## 1. 딕셔너리 생성


```python
dict1 = {
    'name':'홍길동',
    'age':20,
    'address':'서울시 강남구'
}
print(dict1)
```

```python
fruits = {
    'name':'apple',
    'color':'red',
    'emoji':'🍎'
}
print(fruits)
fruits
```

## 2. 아이템 선택 / 추가
- 딕셔너리 key를 이용한 데이터 변경
- 추가/수정 : update()

```python
#딕셔너 객체의 모든 키 리턴
dict1_keys=dict1.keys()
type(dict1_keys)
```

```python
fruits.keys()
```

```python
fruits['emoji']
```

```python
# 딕셔너리 key를 이용한 데이터 변경 
fruits['emoji'] = '🍏'
```

```python
fruits['emoji'] 
```

```python
# 딕셔너리 update함수를 이용한 데이터 변경 - {key:value}
fruits.update({'color':'green'})
fruits
```

```python
# 딕셔너리에 데이터 추가, key가 있으면 update, 없으면 추가
fruits.update({'price':1000})
fruits
```

- copy() : 딕셔너리 객체 복사
- JS의 deep copy 형식으로 독립적인 복사이다. (원본을 해치지 않는 복사)

```python
dict1 = {'korea':'서울','japan':'도쿄'}
dict1
```

```python
dict2= dict1.copy()
dict2
```

```python
print(dict1, dict2)
```

```python
# dict2의 korea 아이템 삭제
```

```python
dict2.pop('korea')
dict2
```

```python
print(dict1,dict2)
```

```python
dict3 = dict1
```

```python
print(dict1,dict3)
```

```python
dict3.pop('korea')
```

```python
print(dict1,dict3) #해당 부분의 경우 shallow copy처럼 원본의 영향을 받는 복사이다
```
