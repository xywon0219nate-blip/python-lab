## 1. 메모리와 철학

#### (1) 메모리 관리 방식 비교

파이썬과 자바스크립트는 모두 자동 메모리 관리를 지원하지만, 내부 구현 방식이 다르다. 
이 차이를 이해하면 대용량 데이터 분석 시 메모리 관리 이슈를 예방할 수 있다.

#### 🔥 핵심 개념 : Everything is an Object!!

파이썬에서는 숫자, 문자열, 함수, 심지어 클래스까지 <span style='color:red'>모두 객체</span>이다. 이 개념은 JS와 가장 크게 다른 철학이다.

#### (2) 비동기 처리 관점 차이


데이터 분석에서 동기 방식이 선호되는 이유 : <span style='color:blue'>전처리</span> -> <span style='color:blue'>분석</span> -> <span style='color:blue'>시각화</span> 순서가 보장되어야 하기 때문이다. 파이썬의 GIL(Global Interpreter Lock)은 CPU 집약적 연산에서 멀티스레드 효과를 제한하므로, 대용량 데이터는 비동기보다 <span style='color:red'>Numpy / Pandas 벡터 연산</span>이 훨씬 효율적이다.



#### (3) Jupyter Notebook 커널 동작 원리

- 커널(Kernel) : 파이썬 인터프리터 프로세스, 노트북을 열면 하나의 커널이 시작된다.
- 셀(Cell) : 코드 실행 단위, 셀은 독립적으로 실행 되지만 변수는 커널 메모리를 공유한다.
- 메모리 공유 : 셀 A에서 정의한 변수를 셀 B에서 바로 사용 가느하다.
- 커널 재시작 : 모든 변수와 임포트가 초기화 된다. 분석 재현을 위해 상단부터 순서대로 실행 필요하다.

📌 Kernel → Restart & Run All 명령어로 전체 재실행하여 '분석 재현성'을 확인하는 습관을 기르자!!!

### 8-1. 람다(lambda) vs 화살표(arrow) 함수
- 공통점은 둘 다 이름 없는 익명 함수(Anonymous function)로 한줄로 작성한다는 것이다.
- 차이점은 화살표 함수는 여러줄 작성이 가능하지만, 람다는 한줄로만 작성 가능하다.

##### ✅ 화살표 함수

---

```js
const add = (x, y) => x + y;          // 한 줄
const add = (x, y) => {               // 여러 줄도 가능
  const result = x + y;
  return result;
}
```

##### ✅ 람다(lambda) 함수

---

```python
add = lambda x, y: x + y             # 한 줄만 가능

# 아래처럼 여러 줄은 불가능! ❌
add = lambda x, y:
    result = x + y                    # SyntaxError
    return result
```



### 8-2. Python에서 lambda 사용법

```python
lst = [3, 1, 4, 1, 5]

# sorted의 key 인자에 짧게 넘기기
lst.sort(key=lambda x: -x)           # 내림차순
print(lst)  # [5, 4, 3, 1, 1]

# map, filter와 함께
doubled = list(map(lambda x: x * 2, lst))
evens   = list(filter(lambda x: x % 2 == 0, lst))

```

- JS 화살표 함수는 **일반 함수의 대체함수**로 사용되고, Python **lambda는 짧은 1회용 함수**로 사용된다.

##### ✅ JavaScript

---

```js
class Animal {
  constructor(name) {
    this.name = name;
  }
  speak() {
    console.log(`${this.name} 소리냄`);
  }
}

class Dog extends Animal {
  speak() {
    console.log(`${this.name} 멍멍`);
  }
}
```

##### ✅ Python

---

```python
class Animal:
    def __init__(self, name):    # constructor → __init__
        self.name = name         # this → self

    def speak(self):             # 모든 메서드에 self 필수
        print(f"{self.name} 소리냄")

class Dog(Animal):               # extends → 괄호 안에 부모 클래스
    def speak(self):
        print(f"{self.name} 멍멍")
        
```
