```python
# '노선명' 컬럼이 가지고 있는 데이터의 종류를 확인, 중복되지 않은 데이터 갯수와 데이터 확인
df['노선명'].nunique() # 27
df['노선명'].unique() # array(['5호선', '4호선',... dtype=object)
```
