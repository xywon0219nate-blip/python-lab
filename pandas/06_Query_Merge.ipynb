## 1. Query를 이용한 데이터 조회
- query() : 조건식을 문자열 형식으로 작성하여 조회

```python
# exam.csv 데이터 로드
exam_df = pd.read_csv('data/exam.csv')
exam_df.head()
```

```python
# exam_df의 형태, 정보 확인
exam_df.shape
exam_df.info()
```

```python
# 수학 점수가 70점 이상인 데이터 조회
# math >= 70
exam_df.query('math >= 70')
```

```python
# 수학 점수가 70점 이상인 2반 학생들 데이터 조회
# 두 개 이상의 조건은 And(&), OR(|) 연산 적용
# math >= 70 & nclass == 2

# exam_df.query('math >= 70 and nclass == 2')
exam_df.query('math >= 70 & nclass == 2')
```

```python
# 1반 이거나 또는 영어 점수가 90점 이상인 데이터 조회
exam_df.query('nclass == 1 or english >= 90')
```

```python
# nclass 기준 groupby
exam_df.groupby('nclass')['id'].count()

# 1, 3, 5반 데이터만 조회
#exam_df.query('nclass == 1 or nclass == 3 or nclass == 5')
#exam_df.query('nclass == 1 | nclass == 3 | nclass == 5')
exam_df.query('nclass in [1, 3, 5]')
```

```python
# 1반의 수학, 과학 점수 평균 조회
exam_df.query('nclass == 1')[['math', 'science']].mean()
```

```python
# 1반의 수학, 과학 점수 중앙값 조회
exam_df.query('nclass == 1')[['math', 'science']].median()
```

```python
# 변수를 이용하여 query 조건 생성
a = int(input('반을 입력해주세요'))
exam_df.query('nclass == @a')
```

```python
# 변수를 이용하여 query 조건 생성
a = int(input('반을 입력해주세요'))
exam_df.query('nclass == @a')[['nclass', 'english']]
```

## 2. Merge
- SQL의 조인과 유사함
- inner join
- left join
- right join
- full outer join

### (1) Inner join
- key를 기준으로 두 개의 데이터프레임(테이블)에 동일하게 존재하는 데이터를 추출
- 예제) '이름'을 키로하여 Inner join(교집합) 수행


```python
df1 = pd.DataFrame(
    {
        '이름' : ['홍길동', '이순신', '김유신'],
        '키' : [180, 170, 175]
    }
)
df1
```

```python
df2 = pd.DataFrame(
    {
        '이름' : ['홍길동', '이순신', '김영희'],
        '성별' : ['남자', '남자', '여자']
    }
)
df2
```

```python
# Inner Join, how = inner 기본값, 생략 가능
pd.merge(df1, df2)
```

### (2) Left Outer Join
- 왼쪽 데이터셋에 있는 데이터프레임(테이블)의 키를 기준으로 병합


```python
# df1, df2 데이터프레임 left Outer Join(Inner Join + Left 테이블 모든 데이터셋)
print(df1)
print('-'*30)
print(df2)
```

```python
# df1, df2 데이터프레임 left Outer Join(Inner Join + Left 테이블 모든 데이터셋)
pd.merge(df1, df2, how ='left')
```

### (3) Right Outer Join
- 오른쪽 데이터프레임에 있는 모든 데이틀 병합


```python
print(df1)
print('-'*30)
print(df2)
```

```python
# Right Outer Join
pd.merge(df1, df2, how='right')
```

### (4) Full Outer Join
- 키를 기준으로 두 데이터프레임의 존재하는 모든 데이터를 기준으로 병합


```python
print(df1)
print('-'*30)
print(df2)
```

```python
# Full Outer Join
pd.merge(df1, df2, how='outer')
```

## 3. Concat 함수를 이용한 데이터 병합


```python
df1 = pd.DataFrame(
    {
        '이름' : ['홍길동', '이순신', '김유신'],
        '키' : [180, 170, 175],
        '나이' : [23, 30, 40],
        '주소' : ['서울', '부산', '원주']
    }
)
df1
```

```python
df2 = pd.DataFrame(
    {
        '이름' : ['Smith', 'Anne', 'Kelly'],
        '키' : [180, 170, 175],
        '나이' : [23, 30, 40],
        '주소' : ['미국', '캐나다', '멕시코']
    }
)
df2
```

```python
# df1, df2를 행 기준으로 concat
df_concat = pd.concat([df1, df2])
df_concat
```

```python
# df_concat 데이터프레임 인덱스 재생성, 기존 인덱스 삭제, 원본 반영
df_concat.reset_index(drop=True, inplace=True)
df_concat
```

```python
# df1, df2를 열 기준으로 concat
pd.concat([df1, df2], axis=1)
```

```python
# df1 객체에 '직업' 컬럼 추가
df1['직업'] = '개발자'
df1
```

```python
pd.concat([df1, df2])
```

```python
pd.concat([df1, df2], join='inner')
```
