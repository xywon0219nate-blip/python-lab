# DataFrame과 리스트, 딕셔너리, ndarray 상호 변환

## 1. 1차원 데이터를 DataFrame 으로 변환

```python
import warnings
warnings.filterwarnings(action='ignore')

import pandas as pd
import numpy as np
```

```python
# 1차원 리스트 생성
col_name1 = ['col_1']
list1 = [1, 2, 3]
print(type(list1), list1)
```

```python
# ndarray로 변환
array1 = np.array(list1)
print(type(array1), array1)
```

```python
# list1, array1를 DataFrame으로 생성(변환)
df1_1 = pd.DataFrame(list1, columns = col_name1, index=['a', 'b', 'c'])
df1_2 = pd.DataFrame(array1, columns = col_name1)
print(df1_1)
print('-'*30)
print(df1_2)
```

## 2. 2차원 데이터를 DataFrame 으로 변환

```python
# 컬럼명 생성
col_names = ['col1', 'col2', 'co3']
col_names
```

```python
# 2차원 리스트 생성
list2 = [[1, 2, 3],
         [4, 5, 6]]
list2
```

```python
# list2를 ndarray로 변환
array2 = np.array(list2)
array2
```

```python
# list2, array2를 DataFrame으로 변환
df2_1 = pd.DataFrame(list2, columns=col_names)
df2_2 = pd.DataFrame(array2, columns=col_names)
print(df2_1, type(df2_1))
print('-'*30)
print(df2_2, type(df2_2))
```

## 3. 딕셔너리를 DataFrame으로 변환

```python
# 딕셔너리 생성
dict1 = {
    'col1': [1, 2, 3],
    'col2': [4, 2, 3],
    'col3': [5, 2, 3]    
}
dict1
```

```python
# DataFrame 생성
df_dict1 = pd.DataFrame(dict1)
df_dict1
```

## 4. DataFrame을 ndarray, list, dictionary로 변환

```python
# df_dict1 데이터프레임을 ndarray로 변환
array3 = df_dict1.values
print(type(array3), '\n', array3)
```

```python
# df_dict1 데이터프레임을 list로 변환, DataFrame.values > Numpy.tolist() > List
list3 = df_dict1.values.tolist()
print(type(list3), '\n', list3)
```

```python
# df_dict1 데이터프레임을 dictionary로 변환
dict3 = df_dict1.to_dict()
print(type(dict3), '\n', dict3)
```

## 5. shape 속성 정리
- python List : ❌ AttributeError
- python Dictionary : ❌ AttributeError
- Numpy ndarray : 사용 가능
- Pandas DataFrame : 사용 가능
- Pandas Series : 사용 가능

```python
# list 생성
print(type(list1), list1)
print(f"shape : {list1.shape}") # ❌ AttributeError: 'list' object has no attribute 'shape'
```

```python
# dictionary 생성
print(type(dict1), dict1)
print(f"shape : {dict1.shape}") # ❌ AttributeError: 'dict' object has no attribute 'shape'
```

```python
# array 
print(type(array1), array1)
print(f"shape : {array1.shape}")
```

```python
# DataFrame
print(type(df_dict1), df_dict1)
print(f"shape : {df_dict1.shape}")
```

## 6. DataFrame 컬럼 데이터 생성 및 수정

#### 타이타닉(Titanic) 데이터 셋 로드

```python
titanic_df = pd.read_csv('data/titanic/train.csv')
titanic_df
```

```python
# 상위 5개의 데이셋 조회
titanic_df.head()
```

```python
# 하위 5개의 데이셋 조회
titanic_df.tail()
```

```python
# 데이터셋 기본정보 조회
titanic_df.info()
```

```python
# 데이터셋 수치 정보 조회
titanic_df.describe()
```

```python
# Age_0 컬럼 추가
# titanic_df.info()
titanic_df['Age_0'] = 1  # 컬럼 ⭕ : 값 업데이트, 컬럼 ❌ : 생성
titanic_df.info()
titanic_df.head() 
```

```python
# Age_by_10 컬럼 추가, Age * 10 값 저장
titanic_df['Age_by_10'] = titanic_df['Age'] * 10
titanic_df.info()
titanic_df.head()
```

```python
# 일부 컬럼 조회
# titanic_df['Age'] # 단일 컬럼 조회시 1차원 Series 반환
titanic_df[['Age', 'Age_0', 'Age_by_10']].head()  # 다중 컬럼 조회 시 2차원 형태
```

```python
# Family_no 컬럼 추가, SibSp 컬럼 + Parch 컬럼 값 합계 + 1
titanic_df.info()
titanic_df['Family_no'] = titanic_df['SibSp'] + titanic_df['Parch'] + 1
titanic_df.info()
```
