# 결손 데이터(결측치) 처리

### 결손데이터(결측치)란?
- 결측치(Missing Data)는 값이 존재하지 않는 데이터를 의미
- 판단스에서는 NaN으로 표현됨

```python
import warnings
warnings.filterwarnings(action='ignore')

import pandas as pd
import numpy as np
```

### 1. 결손 데이터 생성

```python
df = pd.DataFrame(
    {
        'Name': ['Smith','Lee','Kim', 'Park'],
        'Age': [20, np.nan, 30, 40],
        'Score': [90, 85, np.nan, 75],
    }
)
df
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

### 2. 결손 데이터 확인
- isna() = isnull()

```python
df
```

```python
df.isna()
df.isnull()
```

```python
# NaN 값이 있는 데이터만 합계로 계산
print(df.isna().sum())
print(df.isnull().sum())
```

### 3. 결손 데이터 삭제
- dropna() : 결손 데이터가 포함된 행 또는 열 삭제

```python
filter_na = df.dropna()
filter_na
```

```python
print(df)
print('-'*30)
print(filter_na)
```

```python
# 결측치 데이터가 포함된 열 삭제
filter_na_cols = df.dropna(axis=1)
print(df)
print('-'*30)
print(filter_na_cols)
```

```python
# 결측치가 저장된 특정 컬럼만 삭제
df.dropna(axis=0, subset=['Age'])
```

```python
# Addr 컬럼 추가, 값은 모두 결측치(NaN)
df['Addr'] = np.nan
df
```

```python
# 모든 데이터가 결측치인 컬럼 삭제
df.dropna(how='all', axis=1, inplace=True)
df
```

```python
# 하나라도 결측치를 포함한 경우 삭제 how='any', axis=0, inplace=False
df.dropna(how='any', axis=0, inplace=False)
```

### 4. 결손 데이터 대체
- fillna() : 결손 데이터를 특정값으로 대체

```python
# 결손 데이터를 0으로 모두 대체
df.fillna(0)
```

```python
# 결손 데이터를 'Unknown'으로 모두 대체
df_str = df.fillna('Unknown')
df_str.info()
```

```python
# Age 컬럼의 결손 데이터만 0으로 대체
df['Age'] = df['Age'].fillna(0) 
```

```python
df
```

### 5. 평균값으로 대체

```python
df = pd.DataFrame(
    {
        'Name': ['Smith','Lee','Kim', 'Park'],
        'Age': [20, np.nan, 30, 40],
        'Score': [90, 85, np.nan, 75],
    }
)
df
```

```python
# Age 컬럼의 결손 데이터를 평균값으로 대체
age_mean = df['Age'].mean()
age_mean
```

```python
df['Age'] = df['Age'].fillna(age_mean)
df
```

```python
# Score 컬럼의 결측치를 평균값으로 대체
df['Score'] = df['Score'].fillna(df['Score'].mean()) 
df
```

### 6. 최빈값으로 대체
- 가장 많이 등장하는 값으로 대체
- mode() : 가장 많이 등장한 값 반환

```python
df = pd.DataFrame(
    {
        'Name': ['Smith','Lee','Kim', 'Park'],
        'Age': [20, np.nan, 30, 40],
        'Score': [90, 85, np.nan, 75],
    }
)
df
```

```python
# Age 컬럼의 최빈값 조회
df['Age'].mode()[0]
```

```python
# Age 컬럼의 결측치를 최빈값 대체
df['Age'] = df['Age'].fillna(df['Age'].mode()[0])
df
```

### 7. 이전 데이터로 대체
- ffill : forward fill

```python
df = pd.DataFrame(
    {
        'Name': ['Smith','Lee','Kim', 'Park'],
        'Age': [20, np.nan, 30, 40],
        'Score': [90, 85, np.nan, 75],
    }
)
df
```

```python
# Age, Score 컬럼의 결측치를 ffill 함수로 대체
df.ffill()
```

### 8. 다음 데이터로 대체
- bfill : backward fill

```python
# Age, Score 컬럼의 결측치를 다음 값으로 대체
df.bfill()
```

```python
# Age 컬럼을 다음 값으로 대체
df['Age'] = df['Age'].bfill()
df
```

### 9. Titanic 데이터 결측치 처리

```python
# 데이터 로드
titanic_df = pd.read_csv('data/titanic/train.csv')
titanic_df.head()
```

```python
titanic_df.info()
```

```python
# 결측치 확인
#titanic_df.isna().sum()
titanic_df.isnull().sum()
```

```python
# Cabin 컬럼의 687개의 결측 데이터는 '0000'로 대체
titanic_df['Cabin'] = titanic_df['Cabin'].fillna('0000')
```

```python
titanic_df.head()
titanic_df.isna().sum()
```

```python
# Age 컬럼은 평균값을 구해 결측 데이터를 대체
titanic_df['Age'] = titanic_df['Age'].fillna(titanic_df['Age'].mean())
titanic_df.isna().sum()
```

```python
# Embarked의 결측치를 최빈값으로 대체
titanic_df['Embarked'].mode()[0]
titanic_df['Embarked'] = titanic_df['Embarked'].fillna(titanic_df['Embarked'].mode()[0])
```

```python
titanic_df.isna().sum()
```
