#### 이상치 탐지

```python
# [예제 3] 이상치 탐지 (code7_4)
import pandas as pd
import numpy as np
from scipy import stats

df = pd.read_csv('data/iris.csv')
sw = df['SepalWidthCm']

# Z-score 방법 (|z| > 2 기준)
z = np.abs(stats.zscore(sw))
outliers_z = sw[z > 2]
print('Z-score 이상치 (|z|>2):')
print(outliers_z.values)

# IQR 방법
Q1 = 2.800000 #sw.quantile(0.25)
Q3 = 3.300000 #sw.quantile(0.75)
IQR = Q3 - Q1
outliers_iqr = sw[(sw < Q1 - IQR*1.5) | (sw > Q3 + IQR*1.5)]
print(f'\nIQR 이상치 (Q1-1.5×IQR={Q1-1.5*IQR:.2f}, Q3+1.5×IQR={Q3+1.5*IQR:.2f}):')
print(outliers_iqr.values)

# 이상치 제거
clean = sw.loc[~sw.isin(outliers_iqr)]
print(f'\n원본 크기: {len(sw)} → 이상치 제거 후: {len(clean)}')
```

#### 정렬 (sort_values), 순위 (rank)

```python
# [예제 4] 정렬 (sort_values), 순위 (rank) (code7_5, code7_6)
import pandas as pd

df = pd.read_csv('data/iris.csv')
df.head()

# 정렬
print('SepalLengthCm 오름차순 상위 5행:')
print(df.sort_values('SepalLengthCm').head())

print('\nSpecies, SepalWidthCm 복합 정렬 상위 5행:')
print(df.sort_values(['Species', 'SepalWidthCm']).head())

# 순위
print('\nPetalLengthCm 오름차순 순위 상위 5:')
print(df['PetalLengthCm'].rank().astype(int).head())
```

#### concat & merge

```python
df1 = pd.DataFrame([
                        [169, 58, 1.0], 
                        [172, 73, 1.2], 
                        [184, 82, 0.7]
                    ],
                   columns=['height', 'weight', 'eye'])
df2 = pd.DataFrame([
                        [176, 71, 0.8, 'M'], 
                        [169, 62, 0.7, 'F']
                    ],
                   columns=['height', 'weight', 'eye', 'gender'])
```

```python
# 행방향 결합
df12 = pd.concat([df1, df2]).reset_index(drop=True)
df12
```

```python
# merge
df1 = pd.DataFrame([['a', 90], ['b', 80], ['c', 40]], columns=['name', 'kor'])
df2 = pd.DataFrame([['a', 75], ['b', 60], ['d', 90]], columns=['name', 'math'])
print(df1, '\n','-'*30, '\n', df2)
```

```python
# inner join : 공통 이름 기준
df1.merge(df2, on='name')
```

```python
# left outer join : df1 기준
df1.merge(df2, how='left', on='name')
```

```python
# left outer join : df1 기준
pd.merge(df1, df2, how='left', on='name')
```

```python
# outer join : 전체 포함
# pd.merge(df1, df2, how='outer')
df1.merge(df2, how='outer')
```
