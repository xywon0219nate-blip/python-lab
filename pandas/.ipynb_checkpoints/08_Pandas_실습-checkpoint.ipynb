```python
# '승차총승객수'이 50,000명 이상이고, '역명' 이름이 '강남'인 데이터 조회
df2.query('승차총승객수 >= 50000 & 역명 == "강남"')
df2.query('승차총승객수 >= 50000 & 역명 == "강남"').count()
# df2[df2['승차총승객수'] >= 50000 & (df2['역명'] == '강남')].count()
```

```python
# '승차총승객수' 컬럼의 최대값
df['승차총승객수'].max()
```

```python
# '승차총승객수' 컬럼의 최소값
df['승차총승객수'].min()
```

```python
# '승차총승객수' 컬럼의 최대값, '하차총승객수' 컬럼의 최소값을 조회
df.agg(
    {
        '승차총승객수': 'max',
        '하차총승객수': 'min'
    }
)
```

```python
# '승차총승객수','하차총승객수' 컬럼의 최대값을 조회
df[['승차총승객수','하차총승객수']].max()
```

```python
# '승차총승객수', '하차총승객수' 컬럼의 최대값, 최솟값,평균값을 조회
df[['승차총승객수','하차총승객수']].agg(['max','min','mean','median'])
```

## 6. 연산

```python
# '승차총승객수' 컬럼에 10을 더하며, '승차총승객수_10' 컬럼을 추가
df['승차총승객수_10'] = df['승차총승객수'] + 10
```

```python
# '승차총승객수' , '하차총승객수' 결과를 '승하차총승객수' 컬럼으로 생성하여 추가
df['승하차총승객수'] = df['승차총승객수'] - df['하차총승객수']
df
```

```python
# '승하차총승객수'의 평균값, 중앙값
df['승하차총승객수'].mean()
df['승하차총승객수'].median()
df['승하차총승객수'].agg(['mean','median'])
```

```python
# Index 재생성 후 df2로 변경, 기존 인백스 컬럼 유지
df2.reset_index(inplace=True)
```

```python
df2
```

```python
# '노선명' 컬럼의 '2호선' 데이터를 추출하여, df2_rail2 객체 생성
df2_rail2 = df.query('노선명 == "2호선"')
df2_rail2
```

```python
# '노선명' 컬럼의 '2호선'이고 '역명'이 '강남'인 데이터 추출
df2_rail2 = df.query('노선명 == "2호선" and 역명 =="강남"')
df2_rail2
```

```python
# '노선명'컬럼의  '2호선'이고, '역명'이 '강남'인 데이터를 추출하여, df_rail2_gn 객체 생성
df_rail2_gn = df.query('노선명 == "2호선" & 역명 == "강남"')  # 31
df_rail2_gn
```

```python
# df2_rail2 데이터프레임과 df_rail2_gn 데이터프레임을 사칙연산 - 인덱스와 컬럼명이 모두 동일해야 연산이 가능!!
df2_rail2.loc[134:] + df_rail2_gn.loc[134:]
```

```python
# df2_rail2 데이터프레임과 df_rail2_gn 데이터프레임을 사칙연산 - Inner Join(교집합)
# pandas.merge()
pd.merge(df2_rail2, df_rail2_gn)
```

```python
# 타이타닉 데이터셋 로드
td = pd.read_csv('data/titanic/train.csv')
td.head()
```

```python
td.info()
```

```python
td.isna().sum()
```

```python
# Age 컬럼은 평균 값으로 결측치 대체, fillna()
age_mean = td['Age'].mean()
age_mean
td['Age'] = td['Age'].fillna(age_mean)
```

```python
td.isna().sum()
```

```python
# Cabin 컬럼은 삭제
td.drop('Cabin', axis=1, inplace=True)
```

```python
td.info()
```

```python
# Embarked 값은 이전 값으로 대체
td['Embarked'] = td['Embarked'].ffill()
```

```python
td.isna().sum()
```
