```python
# [예제 1] 결측값 확인 (code7_1, code7_2)
import pandas as pd
import numpy as np

df = pd.read_csv('data/iris.csv')

# 인위적으로 결측값 주입
df.iloc[0, 1] = pd.NA
df.iloc[0, 2] = pd.NA
df.iloc[1, 2] = np.nan
df.iloc[2, 3] = None

print('컬럼별 결측값 수:')
print(df.isnull().sum())
print()
print('결측값이 있는 행:')
print(df.loc[df.isnull().sum(axis=1) > 0, :])
```

```python
# [예제 2] 결측값 제거
import pandas as pd
import numpy as np

df = pd.read_csv('data/iris.csv')
df.iloc[0, 1] = pd.NA
df.iloc[1, 2] = np.nan

print('정제 전 행 수:', len(df))
df_clean = df.dropna().reset_index(drop=True)  # 결측값 행 제거 + 인덱스 재설정
print('정제 후 행 수:', len(df_clean))
print(df_clean.head())
```

```python
# [예제 3] 이상치 탐지 (code7_4)
import pandas as pd
import numpy as np
from scipy import stats

df = pd.read_csv('data/iris.csv')
sw = df['SepalWidthCm']

# Z-score 방법 (|z| > 2 기준)
z = np.abs(stats.zscore(sw))
outliers_z = sw[z > 2]
print('Z-score 이상치 (|z|>2):')
print(outliers_z.values)

# IQR 방법
Q1 = sw.quantile(0.25)
Q3 = sw.quantile(0.75)
IQR = Q3 - Q1
outliers_iqr = sw[(sw < Q1 - IQR*1.5) | (sw > Q3 + IQR*1.5)]
print(f'\nIQR 이상치 (Q1-1.5×IQR={Q1-1.5*IQR:.2f}, Q3+1.5×IQR={Q3+1.5*IQR:.2f}):')
print(outliers_iqr.values)

# 이상치 제거
clean = sw.loc[~sw.isin(outliers_iqr)]
print(f'\n원본 크기: {len(sw)} → 이상치 제거 후: {len(clean)}')
```

```python
# [예제 4] 정렬 (sort_values), 순위 (rank) (code7_5, code7_6)
import pandas as pd

df = pd.read_csv('data/iris.csv')
df.head()

# 정렬
print('Sepal_LengthCm 오름차순 상위 5행:')
print(df.sort_values('SepalLengthCm').head())

print('\nSpecies, Sepal_Width 복합 정렬 상위 5행:')
print(df.sort_values(['Species', 'SepalWidthCm']).head())

# 순위
print('\nPetal_Length 오름차순 순위 상위 5:')
print(df['PetalLengthCm'].rank().astype(int).head())
```

```python
# [예제 5] 샘플링 (code7_7)
import pandas as pd

df = pd.read_csv('data/iris.csv')

# 임의 샘플링
df20 = df.sample(n=20, random_state=123)
print('임의 샘플링 20개 품종 분포:')
print(df20['Species'].value_counts())

# 층화 샘플링 (품종별 20%)
stratified = df.groupby('Species').apply(
    lambda x: x.sample(frac=0.2, random_state=123)
)
print('\n층화 샘플링 20% 품종 분포:')
print(stratified['Species'].value_counts())
```

```python
# [예제 7] concat으로 DataFrame 결합 (code7_10)
import pandas as pd

df1 = pd.DataFrame([[169, 58, 1.0], [172, 73, 1.2], [184, 82, 0.7]],
                   columns=['height', 'weight', 'eye'])
df2 = pd.DataFrame([[176, 71, 0.8, 'M'], [169, 62, 0.7, 'F']],
                   columns=['height', 'weight', 'eye', 'gender'])

# 행 방향 결합 (같은 컬럼 구조)
df12 = pd.concat([df1, df2]).reset_index(drop=True)
print('행 방향 concat 결과:')
print(df12)
```

```python
# [예제 8] merge로 JOIN (code7_11)
import pandas as pd

df1 = pd.DataFrame([['a', 90], ['b', 80], ['c', 40]], columns=['name', 'kor'])
df2 = pd.DataFrame([['a', 75], ['b', 60], ['d', 90]], columns=['name', 'math'])

print('inner join (공통 이름만):')
print(df1.merge(df2, on='name'))

print('\nleft join (df1 기준 전체):')
print(df1.merge(df2, how='left', on='name'))

print('\nouter join (전체 포함):')
print(df1.merge(df2, how='outer', on='name'))
```
