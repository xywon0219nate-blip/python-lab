# Anaconda(아나콘다) 삭제

## 1. anaconda-clean 실행




## 2. conda install anaconda-clean



## 3. Proceed? y 선택




## 4. anaconda-clean --yes



## 5. 제어판 → 프로그램 추가/제거 → Anaconda3 제거



## 6. 남은 폴더 삭제

탐색기에서 아래 경로 확인 후 폴더 삭제합니다.

* C:\Users\사용자명\anaconda3
* C:\ProgramData\Anaconda3
* C:\Users\사용자명\.conda
* C:\Users\사용자명\.condarc
* C:\Users\사용자명\.continuum


## 7. 환경 변수 정리

1. 시작 메뉴 → 시스템 환경 변수 편집 검색
2. 환경 변수 클릭
3. 시스템 변수 또는 사용자 변수에서 Path 선택
4. 편집 클릭
5. 아나콘다 관련 경로 모두 선택 후 삭제
